```r
List_of_ISAD_G_Areas <- ISAD_G$ISAD_G_Area

No_of_elements_in_ISAD_G_Areas <- aggregate(x = List_of_ISAD_G_Areas,
                          by = list(ISAD_G_Areas = ISAD_G$ISAD_G_Area),
                          FUN = length)
```
